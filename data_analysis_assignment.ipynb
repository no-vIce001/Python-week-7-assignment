```python
import pandas as pd

import matplotlib.pyplot as plt

data = pd.read_csv('sales_data_sample.csv', encoding = 'latin1')

print(data)
```

```python
data.head()
```

```python
data_types = data.dtypes

print("Data Types:")

print(data_types)
```

```python
missing_values = data.isnull().sum()

print("\nMissing Values:")

print(missing_values)
```

```python
basic_stats = data.describe()

print(basic_stats)
```

```python
mean_values = data.groupby('PRODUCTLINE')['SALES'].mean()

mean_values_sorted = mean_values.sort_values(ascending=False)

print(mean_values_sorted)
```

```python
mean_values = data.groupby('PRODUCTLINE')['PRICEEACH'].mean()

mean_values_sorted = mean_values.sort_values(ascending=False)

print(mean_values_sorted)
```

```python
mean_values = data.groupby('PRODUCTLINE')['QUANTITYORDERED'].mean()

mean_values_sorted = mean_values.sort_values(ascending=False)

print(mean_values_sorted)
```

```python
print('On average, the sales for the Classic Cars Productline are the most while that of the Trains Productline are the least. The Sales amount in descending order is shown above.')

print('On average, the price of the Trucks and Buses Productline is the most while that of the Trains Productline is the least. The Price amount in descending order is shown above.')

print('On average, the quantity of the Productline ordered is almost the same, as shown above, hence the price and sales determine the profitability of the Productline and not the quantity.')
```

```python
# Plotting Total Sales vs Month in 2003

data['ORDERDATE'] = pd.to_datetime(data['ORDERDATE'])
data_2003 = data[data['ORDERDATE'].dt.year == 2003]
monthly_sales = data_2003.groupby(data_2003['ORDERDATE'].dt.to_period('M'))['SALES'].sum().reset_index()
monthly_sales['ORDERDATE'] = monthly_sales['ORDERDATE'].dt.to_timestamp()

plt.plot(monthly_sales['ORDERDATE'], monthly_sales['SALES'])
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Sales Trend per Month in 2003')
plt.show()
```

```python
# Plotting a bar chart for Average Sales by Product

data.groupby('PRODUCTLINE')['SALES'].mean().plot(kind='bar')
plt.xlabel('Productline')
plt.ylabel('Average Sales')
plt.title('Average Sales by Product')
plt.show()
```

```python
# Plotting a histogram for QUANTITYORDERED

data['QUANTITYORDERED'].plot(kind='hist', bins=10)
plt.xlabel('Quantity Ordered')
plt.title('Quantitiy of Product Ordered Distribution')
plt.show()
```

```python
# Plotting a scatter plot of QUANTITYORDERED vs SALES

data.plot(kind='scatter', x='QUANTITYORDERED', y='SALES')
plt.title('Qunatity Ordered vs Sales')
plt.show()
```
